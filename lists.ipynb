```python
#adding of the element
l = ['b','y','e']
l[1]='e'
l=l+['e']
print(l)
```

```python
#append of the element
l = ['b','y','e']
l[1]='e'
l=l+['e']
l.append('y')
print(l)
```

```python
#Accept a list with numbers having 10 elements. Remove the third element and all odd elements.
numbers = []  

for i in range(10):  
    num = int(input("Enter a number: ")) 
    numbers.append(num) 

print(" list:", numbers)
```

```python
del numbers[2]
print("After removing 3rd element:", numbers)
```

```python
new_list = []

for n in numbers:
    if n % 2 == 0:     
        new_list.append(n)

print("After removing odd numbers:", new_list)
```
